# Filter vehicle speeds to remove sudden spikes

```python
# Low-pass filter
def low_pass(seq, alpha):
    seq_filt = np.zeros(len(seq))
    seq_filt[0] = seq[0]
    for jj in range(1, len(seq)):
        seq_filt[jj] = seq_filt[jj-1]+alpha*(seq[jj]-seq_filt[jj-1])

    return(seq_filt)

temp0 = pd.Series({0:0}, dtype=object)
for ii in range(NN):
    temp0[ii] = low_pass(dataset["GM.obd.spd_veh.value"][ii], 0.3)
dataset["GM.obd.spd_veh.value.filt"] = temp0

# Plot random sequences of vehicle speed
plt.figure(figsize=(10,10))
for ii in range(3):
    num = random.randint(0, NN)
    plt.subplot(3, 1, ii+1)
    plt.plot(dataset["GM.obd.spd_veh.value"][num], "k-", label="Raw")
    plt.plot(dataset["GM.obd.spd_veh.value.filt"][num], "r-.", label="Filtered")
    plt.ylabel("Spd [km/h]")
    plt.grid(1)
    plt.title("Sequence {:d}".format(num))
    plt.legend(loc="best")
    if (ii == 2):
        plt.xlabel("Sample")

#plt.figure(figsize=(5,4))
#plt.plot(dataset["GM.obd.spd_veh.value"][0], "b-", label="Raw")
#plt.plot(dataset["GM.obd.spd_veh.value.filt"][0], "r-.", label="Filtered")
#plt.xlabel("Sample")
#plt.ylabel("Spd [km/h]")
#plt.grid(1)
#plt.legend(loc="best")
#plt.tight_layout()
#plt.savefig("../Report/Figures/spd.png")
```

# Resample features to common length for each sequence

```python
def resample(seq, to_length, seq_dist=100):
    ### resample
    # Function for resampling data sequence to a desired length
    # Inputs:
    #   seq         - Array with sequence to be resampled
    #   to_length   - Int. with desired length
    #   seq_dist    - Int. with distance of sequence
    # Outputs:
    #   dist_resamp - Array with distances of resampled sequence
    #   seq_resamp  - Array with resampled sequence
    ###
    # Length of input sequence
    seq_len = seq.shape[0]
    # Downsample if sequence is greater than set length
    if (seq_len > to_length):
        # Randomly choose samples from array to fit set length
        seq = np.random.choice(seq, to_length)
        seq_len = seq.shape[0]
    # Create interpolation function for sequence at distances
    dist_intp = np.linspace(0, seq_dist, seq_len)
    seq_intp = interpolate.interp1d(dist_intp, seq)
    # Upsample if sequence is less than set length
    # by drawing distances from uniform distribution
    dist_new = np.random.uniform(0, dist_intp[-1], to_length-seq_len)
    dist_resamp = sorted(np.concatenate((dist_intp, dist_new)))
    # Interpolate sequence at distances
    seq_resamp = seq_intp(dist_resamp)

    return (dist_resamp, seq_resamp)

def resample_df(df, feats, feats_len={}, seq_dist=None):
    ### resample_df
    # Function for resampling features in data frame
    # Inputs:
    #   df            - Data frame with sequences to resample
    #   feats         - Array with name of features to resample
    #   feats_len     - Dict. with desired length of each feature
    #   seq_dist      - Int. with distance of sequence
    # Outputs:
    #   df            - Data frame with resampled sequences
    #   feats_resamp  - List with names of resampled features
    ###
    feats_resamp = []
    # Filter rows with less than 2 points
    for feat in feats:
        # Replace rows with less than two points with NaNs
        df[feat+'_len']= df[feat].apply(lambda seq: 1 if isinstance(seq, float) else seq.shape[0])
        df.mask(df[feat+'_len']<2, inplace = True)
    # Drop Nans (rows with NaN/len<2) and reset index
    df.dropna(subset=feats, inplace = True)
    df.reset_index(drop=True, inplace = True)
    # Resample features
    for feat in feats:
        print('Resampling feature: ',feat)
        to_length = feats_len[feat]
        new_feats_resampled = ['{0}_d_resampled'.format(feat), '{0}_resampled'.format(feat)]
        df[new_feats_resampled] = df.apply(lambda seq: resample(seq[feat], to_length, seq_dist),
                                        axis=1, result_type="expand")
        feats_resamp.append('{0}_resampled'.format(feat))

    return(df,  feats_resamp)
```

# Partition data

```python
# Create iterable dataset class
class Dataset(data.Dataset):
    # Initialize input and target
    def __init__(self, inputs, targets, iri=None):
        self.inputs = inputs
        self.targets = targets
        if (iri is not None):
            self.iri = iri
    # Method for calculating length of dataset
    def __len__(self):
        return len(self.targets)
    # Method for retrieving input and target at given index
    def __getitem__(self, index):
        X = self.inputs[index]
        y = self.targets[index]

        return(X, y)

def create_datasets(inputs, targets, dataset_class, typ, Ptrain=0.7, Pval=0.15, Ptest=0.15, iri=None):
    ### create_datasets
    # Function for partitioning dataset into train-test-validation
    # Inputs:
    #   inputs          - Array [Nseq, Nsamp] with inputs
    #   targets         - Array [Nseq] with targets
    #   dataset_class   - Dataset class
    #   typ             - String spec. if problem is regression (reg), classification (clf), or autocoder (ae)
    #   Ptrain          - Float with training set partition fraction
    #   Pval            - Float with validation set partition fraction
    #   Ptest           - Float with test set partition fraction
    #   iri             - Array [Nseq] with IRI values (for later validation of classification)
    # Outputs:
    #   training_set    - Iterable object with training set
    #   validation_set  - Iterable object with validation set
    #   test_set        - Iterable object with test set
    ###
    iri_train = None
    iri_val   = None
    iri_test  = None
    # Split sequences into partitions
    # Train, validation, and test
    # Regression
    if (typ.lower() == "reg"):
        # Split for train/validation and test
        [inputs_trainval, inputs_test, targets_trainval, targets_test] =\
            train_test_split(inputs, targets, test_size=Ptest, shuffle=True, random_state=1)
        # Split for train and validation
        [inputs_train, inputs_val, targets_train, targets_val] =\
            train_test_split(inputs_trainval, targets_trainval, test_size=Pval/(Ptrain+Pval),\
                             shuffle=True, random_state=1)
    # Classification
    elif (typ.lower() == "clf"):
        # Split for train/validation and test
        [inputs_trainval, inputs_test, targets_trainval, targets_test] =\
            train_test_split(inputs, targets, test_size=Ptest, shuffle=True, random_state=1, stratify=targets)
        if (iri is not None):
            [_, _, iri_trainval, iri_test] =\
                train_test_split(inputs, iri, test_size=Ptest, shuffle=True, random_state=1, stratify=targets)
        # Split for train and validation
        [inputs_train, inputs_val, targets_train, targets_val] =\
            train_test_split(inputs_trainval, targets_trainval, test_size=Pval/(Ptrain+Pval),\
                             shuffle=True, random_state=1, stratify=targets_trainval)
        if (iri is not None):
            [_, _, iri_train, iri_val] =\
                train_test_split(inputs_trainval, iri_trainval, test_size=Pval/(Ptrain+Pval),\
                                 shuffle=True, random_state=1, stratify=targets_trainval)
    # Autoencoder
    elif (typ.lower() == "ae"):
        # Split for train/validation and test
        [inputs_trainval, inputs_test, targets_trainval, targets_test] =\
            train_test_split(inputs, targets, test_size=Ptest+Pval, shuffle=True, random_state=1, stratify=targets)
        if (iri is not None):
            [_, _, iri_trainval, iri_test] =\
                train_test_split(inputs, iri, test_size=Ptest+Pval, shuffle=True, random_state=1, stratify=targets)
        # Split train and validation using "low" IRI samples for training and "high" for validation
        ind = (targets_trainval == 1).flatten()
        inputs_train  = inputs_trainval[ind==0,:]
        inputs_val    = inputs_trainval[ind==1,:]
        targets_train = targets_trainval[ind==0]
        targets_val   = targets_trainval[ind==1]
        if (iri is not None):
            iri_train = targets_trainval[ind==0]
            iri_val   = targets_trainval[ind==1]

    # Standardize data
    # For time series data
    if (inputs_train.ndim == 2):
        # Average mean and std. of acceleration and speed across sequences
        meanFeat = np.mean(inputs_train, axis=0)
        stdFeat  = np.std(inputs_train, axis=0)
        # Standardize acceleration and speed
        inputs_train = (inputs_train-meanFeat)/stdFeat
        inputs_val   = (inputs_val-meanFeat)/stdFeat
        inputs_test  = (inputs_test-meanFeat)/stdFeat
    # For time series data
    elif (inputs_train.ndim == 3):
        # Average mean and std. of acceleration and speed across sequences
        meanAcc = np.mean( np.mean( inputs_train[:,:,0], axis=1) )
        meanSpd = np.mean( np.mean( inputs_train[:,:,1], axis=1) )
        stdAcc  = np.mean( np.std( inputs_train[:,:,0], axis=1) )
        stdSpd  = np.mean( np.std( inputs_train[:,:,1], axis=1) )
        # Standardize acceleration and speed
        inputs_train[:,:,0] = (inputs_train[:,:,0]-meanAcc)/stdAcc
        inputs_train[:,:,1] = (inputs_train[:,:,1]-meanSpd)/stdSpd
        inputs_val[:,:,0]   = (inputs_val[:,:,0]-meanAcc)/stdAcc
        inputs_val[:,:,1]   = (inputs_val[:,:,1]-meanSpd)/stdSpd
        inputs_test[:,:,0]  = (inputs_test[:,:,0]-meanAcc)/stdAcc
        inputs_test[:,:,1]  = (inputs_test[:,:,1]-meanSpd)/stdSpd
    # Create datasets
    training_set   = dataset_class(inputs_train, targets_train, iri_train)
    validation_set = dataset_class(inputs_val, targets_val, iri_val)
    test_set       = dataset_class(inputs_test, targets_test, iri_test)

    return(training_set, validation_set, test_set)
```

```python
if (runReg == 1):
    # Dense neural network for regression
    class NN_dens_reg(nn.Module):
        def __init__(self, num_feat, N_hid_1, N_hid_2, N_hid_3):
            super(NN_dens_reg, self).__init__()
            self.num_feat = num_feat  # Number of input features
            self.N_hid_1  = N_hid_1   # Number of hidden units in 1st hidden layer
            self.N_hid_2  = N_hid_2   # Number of hidden units in 2nd hidden layer
            self.N_hid_3  = N_hid_3   # Number of hidden units in 3rd hidden layer
            # First hidden layer
            self.hid_1 = nn.Linear(in_features=num_feat,
                                   out_features=N_hid_1,
                                   bias=True)
            self.act_1 = nn.ReLU()
            # Second hidden layer
            self.hid_2 = nn.Linear(in_features=N_hid_1,
                                   out_features=N_hid_2,
                                   bias=True)
            self.act_2 = nn.ReLU()
            # Third hidden layer
            self.hid_3 = nn.Linear(in_features=N_hid_2,
                                   out_features=N_hid_3,
                                   bias=True)
            self.act_3 = nn.ReLU()
            # Output layer
            self.out = nn.Linear(in_features=N_hid_3,
                                 out_features=1,
                                 bias=True)
            # Dropout
            self.drop = torch.nn.Dropout(p=0.25, inplace=False)

        def forward(self, xx):
            # ReLU activation for first hidden layer
            y_hat = self.hid_1(xx)
            y_hat = self.drop(y_hat)
            y_hat = self.act_1(y_hat)
            # ReLU activation for second hidden layer
            y_hat = self.hid_2(y_hat)
            y_hat = self.drop(y_hat)
            y_hat = self.act_2(y_hat)
            # ReLU activation for second hidden layer
            y_hat = self.hid_3(y_hat)
            y_hat = self.drop(y_hat)
            y_hat = self.act_3(y_hat)
            # Output layer without activation
            y_hat = self.out(y_hat)

            return(y_hat)
```

# Supervised learning: CNN regression

```python
# Create convolutional neural network
if (runCNNreg == 1):
    # Function for calculating CNN output sizes
    def cnn_outp(input_length, kernel, padding, stride):
        output_length = ((input_length - kernel + 2*padding)/stride)+1
        output_length = int(np.floor(output_length))

        return(output_length)

    class CNN_reg(nn.Module):
        def __init__(self, len_signal, num_signal,
                     conv_1_chan, conv_1_kern, conv_1_strd, conv_1_padd,
                     conv_2_chan, conv_2_kern, conv_2_strd, conv_2_padd,
                     conv_3_chan, conv_3_kern, conv_3_strd, conv_3_padd,
                     pool_1_kern, pool_1_strd, pool_1_padd,
                     dens_1_outp, dens_2_outp):
            super(CNN_reg, self).__init__()
            print("Inp: ", len_signal)
            # Convolutional layer
            self.conv_1      = nn.Conv1d(num_signal, conv_1_chan, conv_1_kern, conv_1_strd, conv_1_padd)
            self.conv_1_act  = nn.ReLU()
            self.conv_1_outp = cnn_outp(len_signal, conv_1_kern, conv_1_padd, conv_1_strd)
            print("Conv. 1 outp: ", self.conv_1_outp)
            # Convolutional layer
            self.conv_2      = nn.Conv1d(conv_1_chan, conv_2_chan, conv_2_kern, conv_2_strd, conv_2_padd)
            self.conv_2_act  = nn.ReLU()
            self.conv_2_outp = cnn_outp(self.conv_1_outp, conv_2_kern, conv_2_padd, conv_2_strd)
            print("Conv. 2 outp: ", self.conv_2_outp)
            # Max pool layer
            self.pool_1      = nn.MaxPool1d(pool_1_kern, pool_1_strd, pool_1_padd)
            self.pool_1_outp = cnn_outp(self.conv_2_outp, pool_1_kern, pool_1_padd, pool_1_strd)
            print("MaxP. 1 outp: ", self.pool_1_outp)
            # Convolutional layer
            self.conv_3      = nn.Conv1d(conv_2_chan, conv_3_chan, conv_3_kern, conv_3_strd, conv_3_padd)
            self.conv_3_act  = nn.ReLU()
            self.conv_3_outp = conv_3_chan*cnn_outp(self.pool_1_outp, conv_3_kern, conv_3_padd, conv_3_strd)
            print("Conv. 3 outp: ", self.conv_3_outp)
            # Dense layer
            self.dens_1 = nn.Linear(in_features=self.conv_3_outp,
                                    out_features=dens_1_outp,
                                    bias=True)
            self.dens_1_act = nn.ReLU()
            # Dense layer
            self.dens_2 = nn.Linear(in_features=dens_1_outp,
                                    out_features=dens_2_outp,
                                    bias=True)
            self.dens_2_act = nn.ReLU()
            # Output layer
            self.out = nn.Linear(in_features=dens_2_outp,
                                 out_features=1,
                                 bias=True)

        def forward(self, xx):
            # Swap axes to [Nbatch, Nfeat, Lsig]
            xx = torch.swapaxes(xx, -2, -1)
            # Convolutional layer followed by ReLU activation
            yhat = self.conv_1(xx)
            yhat = self.conv_1_act(yhat)
            # Convolutional layer followed by ReLU activation
            yhat = self.conv_2(yhat)
            yhat = self.conv_2_act(yhat)
            # Max pool layer
            yhat = self.pool_1(yhat)
            # Convolutional layer followed by ReLU activation
            yhat = self.conv_3(yhat)
            yhat = self.conv_3_act(yhat)
            # Flatten before dense layer
            yhat = yhat.view(-1, self.conv_3_outp)
            # Dense layer followed by ReLU activation
            yhat = self.dens_1(yhat)
            yhat = self.dens_1_act(yhat)
            # Dense layer followed by ReLU activation
            yhat = self.dens_2(yhat)
            yhat = self.dens_2_act(yhat)
            # Dense output layer
            yhat = self.out(yhat)

            return(yhat)
```

```python
# Create recurrent neural network
if (runRNNreg == 1):
    class RNN_reg(nn.Module):
        def __init__(self, input_size, hidden_size, num_layers, dens_1, dens_2, dens_3):
            super(RNN_reg, self).__init__()
            self.num_layers  = num_layers   # Number of layers
            self.input_size  = input_size   # Input size
            self.hidden_size = hidden_size  # Hidden state size
            self.dens_1      = dens_1       # Dense layer 1 hidden units
            self.dens_2      = dens_2       # Dense layer 2 hidden units
            self.dens_3      = dens_3       # Dense layer 3 hidden units
            # Long short term memory layer
            self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size,
                                num_layers=num_layers, batch_first=True, bias=True)
            # Dense layer 1
            self.dens_1 = nn.Linear(in_features=hidden_size,
                                   out_features=dens_1,
                                   bias=True)
            self.act_1 = nn.ReLU()
            # Dense layer 2
            self.dens_2 = nn.Linear(in_features=dens_1,
                                   out_features=dens_2,
                                   bias=True)
            self.act_2 = nn.ReLU()
            # Dense layer 3
            self.dens_3 = nn.Linear(in_features=dens_2,
                                   out_features=dens_3,
                                   bias=True)
            self.act_3 = nn.ReLU()
            # Output layer
            self.out = nn.Linear(in_features=dens_3,
                                 out_features=1,
                                 bias=True)

        def forward(self, xx):
            # RNN returns output and last hidden state
            y_hat, (hh, cc) = self.lstm(xx)
            # Use only last output from sequences
            #y_hat = y_hat[:,-1,:]
            # Mean over sequences
            y_hat = torch.mean(y_hat, dim=1)
            # Flatten output for feed-forward layer
            y_hat = y_hat.view(-1, self.hidden_size)
            # Dense layer and ReLU activation
            y_hat = self.dens_1(y_hat)
            y_hat = self.act_1(y_hat)
            # Dense layer and ReLU activation
            y_hat = self.dens_2(y_hat)
            y_hat = self.act_2(y_hat)
            # Dense layer and ReLU activation
            y_hat = self.dens_3(y_hat)
            y_hat = self.act_3(y_hat)
            # Output layer
            y_hat = self.out(y_hat)

            return(y_hat)
```

```python
if (runRNNreg == 1):
    # Hyper-parameters
    rnn_reg_num_epochs  = 24       # Number of epochs
    rnn_reg_input_size  = 2        # Input features
    rnn_reg_hidden_size = 20       # Hidden layer size
    rnn_reg_num_layers  = 2        # Number of hidden layers
    rnn_reg_dens_1      = 256      # Dense layer 1 hidden units
    rnn_reg_dens_2      = 128      # Dense layer 2 hidden units
    rnn_reg_dens_3      = 16       # Dense layer 3 hidden units

    # Initialize a new network
    net_rnn_reg = RNN_reg(num_signal, rnn_reg_hidden_size, rnn_reg_num_layers,
                          rnn_reg_dens_1, rnn_reg_dens_2, rnn_reg_dens_3)
    net_rnn_reg.to(device)

    # Define a loss function and optimizer
    rnn_reg_optimizer = optim.Adam(net_rnn_reg.parameters(), lr=0.01)
    rnn_reg_criterion = nn.MSELoss()
    rnn_reg_criterion2 = nn.MSELoss(reduction='none')  # Output loss for every sample in batch for testing

    # Track losses
    rnn_reg_train_loss = []
    rnn_reg_val_loss   = []

    # Train network
    print("Start training...")
    for ii in range(rnn_reg_num_epochs):
        # Epoch loss
        epoch_training_loss = 0
        epoch_validation_loss = 0
        # Begin training
        net_rnn_reg.train()
        for (xx,yy) in train_dataloader_reg_rnn:
            # Transfer to GPU
            xx = xx.to(device)
            yy = yy.to(device)
            # Forward pass
            y_hat = net_rnn_reg(xx.float())
            # Compute loss
            loss = rnn_reg_criterion(y_hat, yy.float())
            # Backward pass
            rnn_reg_optimizer.zero_grad()
            loss.backward()
            rnn_reg_optimizer.step()
            # Update loss
            epoch_training_loss += loss.detach().to("cpu").numpy()
        # Begin validation
        net_rnn_reg.eval()
        for (xx,yy) in val_dataloader_reg_rnn:
            # Transfer to GPU
            xx = xx.to(device)
            yy = yy.to(device)
            # Forward pass
            y_hat = net_rnn_reg(xx.float())
            # Compute loss
            loss = rnn_reg_criterion(y_hat, yy.float())
            # Update loss
            epoch_validation_loss += loss.detach().to("cpu").numpy()
        # Save loss for plot
        rnn_reg_train_loss.append(epoch_training_loss/len(train_dataloader_reg_rnn))
        rnn_reg_val_loss.append(epoch_validation_loss/len(val_dataloader_reg_rnn))
        # Print loss
        if ((ii == 0) or (ii+1) % 1 == 0):
            print("   Epoch {:d} - training loss: {:.5f}, test loss: {:.5f}".format(\
                        ii+1, rnn_reg_train_loss[-1], rnn_reg_val_loss[-1]))

    # Plot training and validation loss
    epoch = np.arange(len(rnn_reg_train_loss))+1
    plt.figure(figsize=(10, 6))
    plt.plot(epoch, rnn_reg_train_loss, "k-", label="Training loss")
    plt.plot(epoch, rnn_reg_val_loss, "r-.", label="Validation loss")
    plt.title("RNN regression training curves")
    plt.xlabel("Epoch")
    plt.ylabel("MSE")
    plt.legend(loc="best")
    plt.grid(1)
```

# Unsupervised learning: autoencoder

```python
if (runAE == 1):
    # Create auto encoder network
    class AutoEncoder(nn.Module):
        def __init__(self, hidden_units_1, hidden_units_2, hidden_units_3, latent_features=2):
            super(AutoEncoder, self).__init__()
            # Encoder
            self.encoder = nn.Sequential(
                # Hidden layer
                nn.Linear(in_features=num_features, out_features=hidden_units_1, bias=True),
                nn.ReLU(),
                # Hidden layer
                nn.Linear(in_features=hidden_units_1, out_features=hidden_units_2, bias=True),
                nn.ReLU(),
                # Hidden layer
                nn.Linear(in_features=hidden_units_2, out_features=hidden_units_3, bias=True),
                nn.ReLU(),
                # Bottleneck layer
                nn.Linear(in_features=hidden_units_3, out_features=latent_features, bias=True)
            )
            # Decoder
            self.decoder = nn.Sequential(
                # Bottleneck layer
                nn.Linear(in_features=latent_features, out_features=hidden_units_3, bias=True),
                nn.ReLU(),
                # Hidden layer
                nn.Linear(in_features=hidden_units_3, out_features=hidden_units_2, bias=True),
                nn.ReLU(),
                # Hidden layer
                nn.Linear(in_features=hidden_units_2, out_features=hidden_units_1, bias=True),
                nn.ReLU(),
                # Output layer
                nn.Linear(in_features=hidden_units_1, out_features=num_features, bias=True),
            )

        def forward(self, xx): 
            zz    = self.encoder(xx)
            x_hat = self.decoder(zz)

            return(zz, x_hat)
```
